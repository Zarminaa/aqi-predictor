```python
import hopsworks

# Connect
project = hopsworks.login()
fs = project.get_feature_store()

# Create a new HUDI Feature Group
fg = fs.get_or_create_feature_group(
    name="aqi_features_hudi",
    version=1,
    description="AQI forecasting features (HUDI)",
    primary_key=["datetime"],
    time_travel_format="HUDI"
)

print("Feature Group:", fg.name)
print("Version:", fg.version)
print("Time travel format:", fg.time_travel_format)
```

```python
import pandas as pd

# Load the engineered features
df = pd.read_csv("features.csv")

# Restore datetime type
df["datetime"] = pd.to_datetime(df["datetime"]).dt.floor("us")

print(df.shape)
df.head()
```

```python
fg.insert(
    df,
    wait=True
)

print("Initial backfill completed!")
```

```python
result = fg.read()

print(result.shape)
result.head()
```
